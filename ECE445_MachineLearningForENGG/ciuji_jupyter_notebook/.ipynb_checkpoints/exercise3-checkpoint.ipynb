# Gradient Descent

```python
from sympy import *
import numpy as np
from numpy.linalg import cholesky
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import display,Latex

i=0
w1,w2,f_w,s_s=symbols('w1 w2 f_w s_s')#s=step_size
W_00=np.array([0,-4])
f_w =(w1**2+ w2-11)**2 + (w1+w2**2-7)**2
k=0.01
iteration=0
```

```python
class func_grad:
    def __init__(self,i,W_0,k):
        X=[]
        Y=[]
        while(i<10000):
            X.append(W_0[0])
            Y.append(W_0[1])
            grad_w1=diff(f_w,w1).subs([(w1,W_0[0]),(w2,W_0[1])])
            grad_w2=diff(f_w,w2).subs([(w1,W_0[0]),(w2,W_0[1])])
            p=-np.array([grad_w1,grad_w2],dtype='float64')

            W_1=p*k+W_0
        #        print (W_1[0],W_1[1],f_w.subs([(w1,W_1[0]),(w2,W_1[1])]).evalf())#look the output        
            if (np.linalg.norm(p)**2<10**-12):
                self.W=(W_1[0],W_1[1])
                self.Z=f_w.subs([(w1,W_1[0]),(w2,W_1[1])]).evalf()
                self.X=X
                self.Y=Y
                self.i=i
                break;
            i+=1
            W_0=W_1
```

```python
display(Latex(r'Set the step size $\gamma$ = 0.005 the value of $\mathbf W^*$ when the algorithm converges:'))
a=func_grad(i,W_00,0.005)
print(a.W)
print("The number of iterations when it took the algorithm to coverge:",a.i)
display(Latex(r'Set the step size $\gamma$ = 0.01 the value of $\mathbf W^*$ when the algorithm converges:'))
a=func_grad(i,W_00,0.01)
print(a.W)
print("The number of iterations when it took the algorithm to coverge:",a.i)
```

Make the stepsize too large and gradient descent will keep on bouncing around a minimizer; conversely, make the stepsize too small and gradient descent will take forever to converge.

In this example, we have the same solution of the two step sizes, it means our choice of step sizes are good. 

In the meanwhile, we can find that the step size would influence the number of iterations to take the algorithm to coveerge, we see that gradient descent with $\gamma = 0.01$ takes less than half the number of iterations to converge when compared with the stepsize of $\gamma = 0.005$.

```python
W_i=[0,-4]
display(Latex(r'Set the step size $\gamma$ = 0.01 and set the the $\mathbf W^0$ = $\mathbf [0\  -4]^T$ value of $\mathbf W^*$ when the algorithm converges:'))
a=func_grad(i,W_i,0.01)
print(a.W)
print("The number of iterations when it took the algorithm to coverge:",a.i)
```

```python
W_i=[0.5,-4]
display(Latex(r'Set the step size $\gamma$ = 0.01 and set the the $\mathbf W^0$ = $\mathbf [0.5\  -4]^T$ value of $\mathbf W^*$ when the algorithm converges:'))
b=func_grad(i,W_i,0.01)
print(b.W)
print("The number of iterations when it took the algorithm to coverge:",b.i)
```

```python
W_i=[0,4]
display(Latex(r'Set the step size $\gamma$ = 0.01 and set the the $\mathbf W^0$ = $\mathbf [0 \ \ 4]^T$ value of $\mathbf W^*$ when the algorithm converges:'))
c=func_grad(i,W_i,0.01)
print(c.W)
print("The number of iterations when it took the algorithm to coverge:",c.i)
```

```python
W_i=[0.5,4]
display(Latex(r'Set the step size $\gamma$ = 0.01 and set the the $\mathbf W^0$ = $\mathbf [0.5\ \ 4]^T$ value of $\mathbf W^*$ when the algorithm converges:'))
d=func_grad(i,W_i,0.01)
print(d.W)
print("The number of iterations when it took the algorithm to coverge:",d.i)
```

All of the case converge but the solutions are differen. This means that the function under consideration does not have a unique global solution. The four returned solutions are therefore either global minima (non-unique) or local minima.

```python
print("in the first case, f(w*)=",a.Z)
print("in the second case, f(w*)=",b.Z)
print("in the third case, f(w*)=",c.Z)
print("in the fourth case, f(w*)=",d.Z)
```

this numbers are all small enough that we can regard them as $\approx 0$. Based on this, we can conclude that all four solutions are in fact global minima.

# Parameter Estimation

```python
from scipy.stats import multivariate_normal
```

```python
mean=np.array([-1,0,4,1,0.5])
cov=2*np.eye(5)
```

```python
def generaD(k):
    mat=np.zeros([5,k])
    for i in range(0,k):
        mulD = multivariate_normal.rvs(mean=mean, cov=cov)
        mat[:,i]=mulD
        mat=np.array(mat)
    return mat
```

```python
Data1=generaD(10)
Data2=generaD(100)
Data3=generaD(1000)
Data4=generaD(10000)
Data5=generaD(100000)
```

```python
Data6_1=generaD(20000)
```

```python
Data6_2=generaD(20000)
```

```python
Data6_3=generaD(20000)
```

```python
Data6_4=generaD(20000)
```

```python
Data6_5=generaD(20000)
```

```python
Data6=np.concatenate((Data6_1,Data6_2,Data6_3,Data6_4,Data6_5),axis=1)
Data6.shape
```

```python
mean_hat1=np.mean(Data1,axis=1)
mean_hat2=np.mean(Data2,axis=1)
mean_hat3=np.mean(Data3,axis=1)
mean_hat4=np.mean(Data4,axis=1)
mean_hat5=np.mean(Data5,axis=1)
mean_hat6=np.mean(Data6,axis=1)
```

```python
display(Latex("when N=1,the $ \hat \mu_1 = $"));print(mean_hat1)
display(Latex("when N=1,the $ \hat \mu_2 = $"));print(mean_hat2)
display(Latex("when N=1,the $ \hat \mu_3 = $"));print(mean_hat3)
display(Latex("when N=1,the $ \hat \mu_4 = $"));print(mean_hat4)
display(Latex("when N=1,the $ \hat \mu_5 = $"));print(mean_hat5)
display(Latex("when N=1,the $ \hat \mu_6 = $"));print(mean_hat6)
```

```python
error1=np.linalg.norm(mean-mean_hat1)**2
error2=np.linalg.norm(mean-mean_hat2)**2
error3=np.linalg.norm(mean-mean_hat3)**2
error4=np.linalg.norm(mean-mean_hat4)**2
error5=np.linalg.norm(mean-mean_hat5)**2
error6=np.linalg.norm(mean-mean_hat6)**2
error_arr=[error1,error2,error3,error4,error5,error6]
display(Latex("when N=10,the instantneous error of $ \hat \mu_1$ :   $e_1 = $"),error1)
display(Latex("when N=100,the instantneous error of $ \hat \mu_2$ :  $e_2 = $"),error2)
display(Latex("when N=1000,the instantneous error of $ \hat \mu_3$ :  $e_3 = $"),error3)
display(Latex("when N=10000,the instantneous error of $ \hat \mu_4$ :  $e_4 = $"),error4)
display(Latex("when N=100000,the instantneous error of $ \hat \mu_5$ :  $e_5 = $"),error5)
display(Latex("when N=1000000,the instantneous error of $ \hat \mu_6$ :  $e_6 = $"),error6)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
plt.title("log-log figure of error_N and N")
#for i in range(0,5):
#    line, =ax.plot(error_arr)
ax.plot([10,100,10**3,10**4,10**5,10**6],error_arr,'-*')
ax.set_yscale("log")
ax.set_xscale("log")
plt.xlabel('Number of samples (N)')
plt.ylabel('Instantaneous estimation error')
```

From the plot, we can find that the relationship between $N$ and $e_N$ on a $\log$-$\log$ scale is almost linear with a negative slope, the N increase and the error descrease. That is, $\log(e_N) \propto -\log(N)$. This implies that $e_N \propto N^{-1}$; that is, the instantaneous error seems to be inversely proportional to the number of samples used to obtain the estimate of the mean of a random vector.

So we can conclude that the more sample we have, the best solution we get.
