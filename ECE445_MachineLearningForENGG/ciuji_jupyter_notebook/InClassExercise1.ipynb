```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random


df_data=pd.DataFrame(load_boston().data)
raw_data,raw_amount=load_boston(return_X_y=True)
def randomData():
    indexs=list(range(506))
    random.shuffle(indexs)
    data1=raw_data[indexs[:101]]
    y1=raw_amount[indexs[:101]]
    
    data2=raw_data[indexs[101:202]]
    y2=raw_amount[indexs[101:202]]
    
    data3=raw_data[indexs[202:303]]
    y3=raw_amount[indexs[202:303]]
    
    data4=raw_data[indexs[303:404]]
    y4=raw_amount[indexs[303:404]]
    
    data5=raw_data[indexs[404:506]]
    y5=raw_amount[indexs[404:506]]
    
    
    folds_data=[data1,data2,data3,data4,data5]
    folds_labels=[y1,y2,y3,y4,y5]
    return folds_data,folds_labels


        
fold_number=np.array([0,1,2,3,4])
```

```python
class LSRegression():
    def __init__(self,data,amount):
        self.X=data.astype(np.float64)
        self.Y=amount
        self.XT=self.X.T
        self.XXT=np.mat(self.XT@self.X).astype(np.float64)
        self.WLS=np.mat(self.XXT.I@self.XT@self.Y)
        
    def getWLS(self):
        return self.WLS
    
    
    def fit(self,testData,realY):
        predictY=(testData@self.WLS.T).reshape(realY.shape)
        error=np.linalg.norm(predictY-realY,ord=2)        
        self.error=error
        return(predictY,error)
```

```python
folds_data,folds_labels=randomData()
LSerror=[]
for i in range(5):
    test_data=folds_data[i]
    test_labels=folds_labels[i]
    train_number=np.delete(fold_number,i,axis=0)
    train_data=np.concatenate((folds_data[train_number[0]],folds_data[train_number[1]],folds_data[train_number[2]],folds_data[train_number[3]]))
    train_labels=np.concatenate((folds_labels[train_number[0]],folds_labels[train_number[1]],folds_labels[train_number[2]],folds_labels[train_number[3]]))

    myLS=LSRegression(train_data,train_labels)
    predictY,predictError=myLS.fit(test_data,test_labels)
    plt.title("Measured versus Predicted")
    plt.scatter(test_labels,list(predictY),c='b')
    plt.plot(test_labels,test_labels,c='r')
    plt.xlabel("real value")
    plt.ylabel("predict value")
    LSerror.append(predictError)
    #print(myLS.getWLS())
    print("prediction error of fold ",i," : ",predictError)
    
print("average LS error:",np.mean(LSerror))
```

```python
#%% Ridge regression    
class RRegression():
    def __init__(self,data,amount,regItem=0):
        self.X=data.astype(np.float64)
        self.Y=amount
        self.XT=self.X.T
        self.XXT=np.mat(self.XT@self.X).astype(np.float64)
        regMat=np.eye(13)*regItem
        
        self.WLS=np.mat((self.XXT+regMat).I@self.XT@self.Y)
        
    def getWLS(self):
        return self.WLS
    
    def setWLS(self,userWRR):
        self.WLS=userWRR
        
        
    def fit(self,testData,realY):
        predictY=(testData@self.WLS.T).reshape(realY.shape)
        error=np.linalg.norm(predictY-realY,ord=2)
        #error=list(error)
        
        self.error=error
        return(predictY,error)
        #self.error=sum((predictY-realY)**2)
        #return (self.error)
        
```
