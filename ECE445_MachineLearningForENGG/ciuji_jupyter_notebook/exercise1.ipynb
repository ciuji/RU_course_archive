```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Year = [1930,1940,1950,1960,1970,1980,1990,2000,2010]
```

```python
GDP = [1.015,1.33,2.29,3.26,4.951,6.759,9.366,13.131,15.599]
```

```python
%matplotlib inline
plt.figure()
plt.plot(Year,GDP,"r--.",linewidth=1)
plt.xlabel("Year")
plt.ylabel("Trillions of $")
plt.title("Real US GDP (in trillions)")
```

So I use least squares method to stimulate the curve.
1. Change the x-axis to 0 to 80.
1. I set the largest power of the x to 1.
1. Build a least squares model.
1. Draw the line.
1. I get a straight line. y = 0.1856x - 259.029

```python
%matplotlib inline
x=np.array(Year)-1930
y=np.array(GDP)
m=[]

for i in range(2):
    a=x**(i)
    m.append(a)
A=np.array(m).T
b=y.reshape(y.shape[0],1)

def projection(A,b):
    AA = A.T.dot(A)
    w=np.linalg.inv(AA).dot(A.T).dot(b)
    print(w)
    return A.dot(w)

yw = projection(A,b)
yw.shape=(yw.shape[0],)
plt.figure()
plt.plot(Year,GDP,"r--.",linewidth=1)
plt.xlabel("Year")
plt.ylabel("Trillions of $")
plt.title("Real US GDP (in trillions)")
plt.plot(1930+x,yw)
```

Then I try to set the largest power of x to 2 and I got a good curve to show the relationship between GDP and Year.

The function of the curve is : y = 1.04409091 + 0.00917554 \* (x-1930) + 0.00220593 * (x-1930)^2
