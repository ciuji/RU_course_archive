```python
import numpy as np
from numpy.linalg import cholesky
import matplotlib.pyplot as plt

sampleNo = 1000
Lambda1=np.array([[4,0],[0,1]])
Lambda2=np.array([[1,0],[0,5]])
U=np.array([[1/(2**0.5),-1/(2**0.5)],[1/(2**0.5),1/(2**0.5)]])
Sigma1=U@Lambda1@U.T
Sigma2=U@Lambda2@U.T
Sigma3=[[1,0],[0,1]]
Mu1=[1,0]
Mu2=[-1,0]
Mu3=[0,-10]
```

```python
s1 = np.random.multivariate_normal( Mu1,Sigma1,sampleNo) 
s2 = np.random.multivariate_normal( Mu2,Sigma2,sampleNo) 
s3 = np.random.multivariate_normal( Mu3,Sigma3,sampleNo) 

plt.scatter(s1[:,0],s1[:,1],c='r',label='N1')
plt.scatter(s2[:,0],s2[:,1],c='g',label='N2')
plt.scatter(s3[:,0],s3[:,1],c='b',label='N3')
plt.legend()
plt.show()
```

```python
samples=np.concatenate((s1,s2,s3), axis=0)
```

```python
a=list(range(3000))
random_index=np.random.choice(a, size=3000, replace=False)
c1=[]
c3=[]
c2=[]
for i in random_index[:1000]:
    c1.append(samples[i])
for i in random_index[1000:2000]:
    c2.append(samples[i])
for i in random_index[2000:3000]:
    c3.append(samples[i])
c1=np.array(c1)
c2=np.array(c2)
c3=np.array(c3)
```

```python
plt.scatter(c1[:,0],c1[:,1],c='r',label='random c1')
plt.scatter(c2[:,0],c2[:,1],c='g',label='random c2')
plt.scatter(c3[:,0],c3[:,1],c='b',label='random c3')
plt.legend()
plt.title("random split data into 3 cluster")
plt.show()
```

```python
#%% function part
def calc_means(data1,data2,data3):
    res=[data1.mean(axis=0),data2.mean(axis=0),data3.mean(axis=0)]
    return res
def calc_dist(point,means_arr):
    d1=np.linalg.norm(point-means_arr[0])
    d2=np.linalg.norm(point-means_arr[1])
    d3=np.linalg.norm(point-means_arr[2])
    res=[d1,d2,d3]
    #print (res)
    return res.index(min(res))
```

```python
a=list(range(3000))
random_index=np.random.choice(a, size=3000, replace=False)
c1=[]
c3=[]
c2=[]
for i in random_index[:1000]:
    c1.append(samples[i])
for i in random_index[1000:2000]:
    c2.append(samples[i])
for i in random_index[2000:3000]:
    c3.append(samples[i])
c1=np.array(c1)
c2=np.array(c2)
c3=np.array(c3)
for j in range(50):
    allData=np.concatenate((c1,c2,c3), axis=0)
    newC1=[]
    newC2=[]
    newC3=[]
    means_arr=calc_means(c1,c2,c3)
    for i in allData:
        res=calc_dist(i,means_arr)
        if(res==0):
            newC1.append(i)
        elif(res==1):
            newC2.append(i)
        elif(res==2):
            newC3.append(i)
    if (c1.size/2-len(newC1)==0 and c2.size/2-len(newC2)==0 and c3.size/2-len(newC3)==0):
    #if (list(c1)==newC1 and list(c2)==newC2 and list(c3)==newC3):
        if((np.array(newC1)==c1).all() and (np.array(newC2)==c2).all() and (np.array(newC3)==c3).all()):
            print ("Number of iterations:",j)
            break
    c1=np.array(newC1)
    c2=np.array(newC2)
    c3=np.array(newC3)

plt.scatter(c1[:,0],c1[:,1],c='r',label='final c1')
plt.scatter(c2[:,0],c2[:,1],c='g',label='final c2')
plt.scatter(c3[:,0],c3[:,1],c='b',label='final c3')
plt.legend()
plt.title("clusters after K-Means")
plt.show()
```
